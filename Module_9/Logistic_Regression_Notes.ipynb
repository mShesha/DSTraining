**Logistic Regression**, which is a supervised classification model. It will help you make predictions in cases where the output is a categorical variable.

Since logistic regression is the most easily interpretable of all classification models, it is very commonly used in various industries such as banking, healthcare, etc. For example you want to predict 

* if someone is going to default on their loan payments
* if someone is diabetic or not 
* if an email is spam or not

#### Binary Classification

The most common use of logistic regression models is in binary classification problems.

A classification problem where we have two possible outputs is called a binary classification problem. All the above three examples are binary classification problem.

#### Multiclass Classification

A classification problem where we have more than 2 possible outputs.

**Example:** sorting your emails as promotional/personal/work emails etc.

#### Sigmoid Curve

One way of overcoming this problem of having a very sharp boundary is to think in terms of probabilities. Where the sugar levels are low we expect the probability to be very low and where the sugar levels are high the probabilities to be very high.

**Sigmoid curve** has all the properties you would want — extremely low values in the start, extremely high values in the end, and intermediate values in the middle — it’s a good choice for modelling the value of the probability of diabetes.
